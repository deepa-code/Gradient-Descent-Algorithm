```python
#Goal :Write code Gradient Descent Algorithm without using any package/library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')

#Step 1: Data Preparation

#Download the dataset and Read Data from File 
data = pd.read_csv(r'C:\Users\Sundar\Documents\Deepa\ML\Dataset2\Dataset\Training\Features_Variant_3.csv', header = None) #read from dataset

#Feature Selection
X_file = data.iloc[:,[0,1,2,3,4,29,30,31,32,33,34,35,36,38,39,44,45]]
y_file = data.iloc[:,53] 

#Partition it randomly into train and test set using a 70/30 split.
X_train,X_test,y_train,y_test=train_test_split(X_file,y_file,test_size=0.3,random_state=23)
#print(len(y_train),len(X_train),len(X_test))

#Feature Scaling/Normalisation
X_test = (X_test - np.mean(X_test))/np.std(X_test)
X_train = (X_train - np.mean(X_train))/np.std(X_train)

#Initialize Parameters
beta = np.zeros((18,1))
alpha = 0.1
num_iters = 1000
threshold = 0.01 
iterCosts=[]
alphas=[]
costs=[]
thresholds=[]
iterations=[]
```

```python
#Step 2:Cost Function calculation for Multiple Variables
def computeCostMulti(X, y, beta):
    m = len(y)
    J_prev=np.sum(np.power(np.dot(X, beta) - y, 2)) / (2*m)
    return J_prev

class Result:
  def __init__(self, costs=[], alphas=[],thresholds=[],iteration=[]):
    self.costs = costs
    self.alphas = alphas
    self.thresholds = thresholds
    self.iteration=iteration


#Step 3: Implement the gradient descent algorithm with batch update rule. Use the same cost function as in the class (sum of squared error). 
def gradientDescentMulti(X, y, beta, alpha, num_iters,threshold):
    iters=0
    m = len(y)
    ones = np.ones((m,1))
    X = np.hstack((ones, X))
    y = y[:,np.newaxis]
    J_prev=computeCostMulti(X, y, beta)
    gradient=abs(J_prev)
    while gradient > threshold and iters<num_iters:    
            beta = beta - (alpha/m) * np.dot(X.T, np.dot(X, beta) - y)
            J_cur = np.sum(np.power(np.dot(X, beta) - y, 2)) / (2*m)
            gradient=abs(J_prev-J_cur)
            J_prev=J_cur
            iterCosts.append(J_cur)
            iters = iters+1  #iteration count
    alphas.append(alpha)
    costs.append(J_cur)
    thresholds.append(threshold)
    iterations.append(iters-1)
    print("Iteration",iters,"\tAlpha",alpha,"\tCost is",J_cur,"\tThreshold is",threshold)
    return(Result(costs,alphas,thresholds,iterations))
```

```python
#Step 4: Experiment with various values of learning rate ∝ to see how the error varies for train and test sets with varying ∝. 
def findBestAlpha(X_data,y_data,ds):
    #clear arrays 
    alphas=[]
    costs=[]
    for alpha in np.arange(0.1, 1, 0.1):
        res=gradientDescentMulti(X_data, y_data, beta, alpha, num_iters,threshold) 
    
    #Plot the results. 
    plt.plot(res.alphas,res.costs,label=ds)
    plt.xlabel("Alphas")
    plt.ylabel("Cost")
    plt.title("Graph of Cost for various Alphas of "+ds)
    plt.show()        

#Call the function

#train data
findBestAlpha(X_train,y_train,'Training Data')

#test data
findBestAlpha(X_test,y_test,'Test Data')

#Best Alpha is 0.5 since it has got the least cost or error  for training data and test data
```

```python
#Step 5: Experiment with various thresholds for convergence.
threshs=[1,0.1,0.01,0.001,0.0001,0.00001]
   
def findBestThreshold(X_data, y_data, ds):
    alpha=0.5  #Best Alphas from Step 4
    
    for thresh in threshs:
        threshold=thresh
        res=gradientDescentMulti(X_data, y_data, beta, alpha, num_iters,threshold)
    #print (res.alphas,res.costs ,res.thresholds)
    
    #Plot the results
    plt.plot(res.thresholds,res.costs,label=ds)
    plt.xlabel("thresholds")
    plt.ylabel("Cost")
    plt.title("Graph of Cost for various thresholds of "+ds)
    plt.show() 
    return res

#Call the function

#train data
#clear arrays 
alphas=[]
costs=[]
thresholds=[]
iterations=[]
trainRes=findBestThreshold(X_train,y_train,'Training Data')
trainCosts=trainRes.costs

#test data
#clear arrays 
alphas=[]
costs=[]
thresholds=[]
iterations=[]
testRes=findBestThreshold(X_test,y_test,'Test Data')
testCosts=testRes.costs

#Best Threshold is 0.00001 since it has got the least cost or error  for training data and test data


```

```python
#Check how number of iterations affect cost .Plot Cost Vs Iterations
plt.plot(trainCosts,'r',label='TrainSet')
plt.plot(testCosts,'b',label='TestSet')
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.legend(loc='upper right')
plt.title("Graph of Cost Vs Iterations")
plt.show()
```
